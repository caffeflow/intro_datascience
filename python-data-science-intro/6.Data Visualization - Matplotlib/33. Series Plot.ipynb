# Series可以直接调用Plot方法

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
s = Series(np.random.randn(1000))
```

```python
s.plot(kind = 'line',color='b')
```

```python
s.plot(kind = 'kde',color='b')
```

```python
s.plot(kind = 'hist',color='b')
```

#### Series.cumsum() 

```python
# Return cumulative sum over a DataFrame or Series axis.
t1 = Series(np.arange(1,100,2))
t2 = t1.cumsum()
```

```python
fig,ax = plt.subplots(2,1)
t1.plot(ax=ax[0],label='t1')
ax[0].legend()
t2.plot(ax=ax[1],label='t2=t1.cumsum()')
ax[1].legend()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind='line',label ='s1')
s2.plot(kind='line',label = 's2')
plt.legend()
```

```python
fig,ax = plt.subplots(2,1)
s1.plot(ax=ax[0],color='r',label='s1',title='s1_title')
ax[0].legend()
s2.plot(ax=ax[1],label = 's2',title='s2_title')
ax[1].legend()
# give specified padding.
plt.tight_layout()
```
