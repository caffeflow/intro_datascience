# DataFrame 直接调用Plot方法

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = DataFrame(np.random.randint(1,10,size=40).reshape(10,4), columns=['A','B','C','D'])
```

```python
df
```

```python
# 默认绘图，通过行去绘图，四条线分别代表'A'~'D'四个column
df.plot()
```

```python
# 按照列index去绘图
# 通过for循环index
for i in df.index:
    df.loc[i].plot(label = str(i))
plt.legend()
```

#### 柱形图

```python
df.plot(kind='bar')
```

```python
# barh 水平柱形图
df.plot(kind='barh')
```

```python
# 叠加柱形图
df.plot(kind='bar',stacked=True)
```

```python
df.plot(kind='area')
```
