```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Return evenly spaced numbers over a specified interval.
x = np.linspace(0.0,5.0)
```

```python
x
```

```python
len(x)
```

```python
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*2*x)
```

```python
plt.plot(x,y1,'r--',label = 'sin(pi*x)')
plt.plot(x,y2,'b--',label = 'sin(pi*x*2)')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title("this is x-y value")
# 显示标签
plt.legend()
```

### Subplot ，子图

```python
# 语句放在同一个cell中
# Add a subplot to the current figure.
# Signature: plt.subplot(*args, **kwargs)
# 参数解释，xyn生成x行y列的子图，画第n张图
plt.subplot(221)
plt.plot(x,y1,'b--')
plt.subplot(222)
plt.plot(x,y2,'r--')
plt.subplot(223)
plt.plot(x,y1,'r*')
plt.subplot(224)
plt.plot(x,y2,'b*')
```

### subplots

```python
# 注意，这些语句，要放在一个cell里运行。
# 注意subplots复数，这里的参数写法稍微区别
# 得到的ax就是执行画图的对象
figure,ax = plt.subplots(2,2)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
ax[1][0].plot(x,y1,'b--')
ax[1][1].plot(x,y2,'r--')
plt.show()
```
