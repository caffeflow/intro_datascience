# 对股票数据,进行可视化分析

```python
# basic 
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

# get data
import pandas_datareader as pdr

# virsual
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#time
from datetime import datetime
```

```python
start = datetime(2015,9,20)
alibaba = pdr.get_data_yahoo('BABA',start = start)
amazon = pdr.get_data_yahoo('AMZN',start = start)
```

```python
alibaba.head()
```

```python
amazon.head()
```

```python
# DateFrame Plot
# 每日关闭股价
alibaba['Adj Close'].plot(label = 'alibaba',legend = True)
amazon['Adj Close'].plot(label = 'amazon', legend = True)
```

```python
# 股票交易量 ， 亚马逊交易价格更高
alibaba['Volume'].plot(label = 'alibaba',legend = True)
amazon['Volume'].plot(label = 'amazon',legend = True)
```

```python
# 交易价格起伏  ,可以看到亚马逊的交易价格起伏更大
alibaba['High-Low'] = alibaba['High'] - alibaba['Low']
amazon['High-Low'] = amazon['High'] - amazon['Low']

alibaba['High-Low'].plot(label = 'alibaba',legend = True)
amazon['High-Low'].plot(label = 'amazon',legend = True)
```

```python
# daliy return 每天之间的涨跌
# Series.pct_change() ：Percentage change between the current and a prior element.
alibaba['daliy return'] = alibaba['Adj Close'].pct_change()
alibaba['daliy return'].head()
```

```python
# marker: 取值的点 
alibaba['daliy return'].plot(label = 'alibaba',figsize=(10,6),marker='o',linestyle ='--',legend = True)
```

```python
# 直方图
# marker: 取值的点 
alibaba['daliy return'].plot(kind='hist',label = 'alibaba',figsize=(10,6),legend = True)
```

```python
# 正太分布图
alibaba['daliy return'].plot(kind = 'kde')
```
