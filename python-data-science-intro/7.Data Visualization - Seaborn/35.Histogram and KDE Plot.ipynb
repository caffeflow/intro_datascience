# seaborn 基于matplotlib
 这里使用seaborn实现直方图和密度图


```python
# basic
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

#matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

#seaborn
import seaborn as sns
```

```python
s = Series(np.random.randn(1000))
```

### 回顾matplotlib, Series plot

```python
plt.hist(s)
```

```python
s.plot(kind= 'hist')
```

```python
s.plot(kind = 'kde')
```

### seabon

```python
# 默认直方图和密度图
sns.distplot(s)
```

```python
# bins : argument for matplotlib hist(), or None, optional
# kde=False 不画kde
sns.distplot(s,kde=False,color ='r',bins =100)
```

```python
# kde专用
# shade : draw with filled contours when data is bivariate
sns.kdeplot(s,color='r',shade='True')
```
