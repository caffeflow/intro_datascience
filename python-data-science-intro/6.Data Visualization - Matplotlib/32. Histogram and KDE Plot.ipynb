# 画直方图和密度图

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

```python
# randn随机正太分布
s = Series(np.random.randn(1000))
```

```python
# 两个array,第一个是分布array,第二个是数据取值范围
plt.hist(s)
```

```python
t = plt.hist(s,rwidth=0.9)
```

#### 特殊例子，均匀分布

```python
s1 = Series(np.arange(10))
```

```python
s1
```

```python
plt.hist(s1,rwidth=0.9)
```

#### 指定区间精度

```python
plt.hist(s,rwidth=0.9,bins=50,color='r')
plt.show()
```
