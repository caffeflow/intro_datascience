# 对时间序列进行采样和画图

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
import datetime
```

```python
t_range = pd.date_range('2016-01-01','2016-12-31')
```

```python
t_range
```

```python
s = Series(np.random.rand(len(t_range)),index = t_range)
```

```python
s.head()
```

```python
s['2016-1'].mean()
```

### 采样 

```python
# 按月新生成数据, 月份平均数  
s_month = s.resample('M').mean()
```

```python
s_month
```

### 填充缺失值，之前以天为单位，现在用小时

```python
s_hour= s.resample('H')
s_hour
```

```python
# 可见，1月1日缺失的小时数据，由1月2日来填充，类推
s_hour= s.resample('H').bfill()
s_hour
```
