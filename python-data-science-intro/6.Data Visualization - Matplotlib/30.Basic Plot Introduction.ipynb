# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot y versus x as lines and/or markers.
a = [1,6,3]
b = [4,5,6]
# (1,4)(6,5)(3,6)  线条格式统统可以设置
plt.plot(a,b,'r*')
plt.show()
```

```python
# matplotlib inline运行后，可直接绘制图，而不用再plt.show()
%matplotlib inline 
```

```python
plt.plot(a,b)
```

```python
# 变为虚线
plt.plot(a,b,'--')
```

```python
# 同一画布中画两个图
c = [10,8,6]
d = [1,8,3]
plt.plot(a,b,'--',c,d,'-')
```

### 实例

```python
# 横坐标点
t = np.arange(0.0,2.0,0.1)
# 纵坐标点
s = np.sin(t*np.pi)
# 绘图，red虚线
plt.plot(t,s,'r--',label = "aaaa")
# 绘图，blue虚线
plt.plot(t*2,s,'b--',label = 'bbbb')
# 横、纵坐标标签
plt.xlabel('this is x')
plt.ylabel('this is y')
# 图表标题
plt.title('this is a demo')
# 将label = 'aaaa' 这类标签，作为图例显示在右上角。
plt.legend()
```
